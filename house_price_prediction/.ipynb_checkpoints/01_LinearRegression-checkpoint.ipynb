```python
import numpy as np
from sklearn.impute import SimpleImputer

# Create an imputer object to replace NaN values with the mean of each column
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# Fit and transform the training data
X_train_imputed = imputer.fit_transform(X_train)

# Check if Y_train has NaN values and impute if necessary
# First, ensure Y_train is a numpy array
Y_train = np.array(Y_train)

# Check if Y_train contains numeric data that can be checked for NaN
if np.issubdtype(Y_train.dtype, np.number):  # Check if Y_train contains numeric data
    if np.isnan(Y_train).any():
        # For regression tasks, you can impute Y_train with mean
        y_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
        Y_train = y_imputer.fit_transform(Y_train.reshape(-1, 1)).ravel()
        # For classification tasks, you might want to use 'most_frequent' strategy instead
else:
    # If Y_train is not numeric, handle accordingly (e.g., for categorical data)
    # For example, you might use SimpleImputer with strategy='most_frequent'
    y_imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    Y_train = y_imputer.fit_transform(Y_train.reshape(-1, 1)).ravel()

# Training the model with the cleaned train set
model.fit(X_train_imputed, Y_train)
```

```python
# Training Error
from sklearn.metrics import mean_squared_error

Y_train_pred = model.predict(X_train)

train_error = np.sqrt(mean_squared_error(Y_train, Y_train_pred))

print("Mean squared error: %.3f" % train_error)
```

```python
#Test Error
Y_test_pred = model.predict(X_test)

test_error = np.sqrt(mean_squared_error(Y_test, Y_test_pred))

print("Mean squared error: %.3f" % test_error)
```
